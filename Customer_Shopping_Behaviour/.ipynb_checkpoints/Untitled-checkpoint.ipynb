```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

#### Missing values in Review Rating were imputed using the median of their respective Category to preserve category-specific rating patterns and reduce the impact of outliers.

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df
```

```python
labels=['Young Adult','Adult','Middle-aged','Senior']
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
# Purchase_Frequency_days
frequency_of_purchases={
    "Every 3 Months":90,
    "Annually":365,
    "Quarterly":90,
    "Monthly":30,
    "Bi-Weekly":14,
    "Fortnightly":14,
    "Weekly":7
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_of_purchases)
```
